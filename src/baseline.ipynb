# Baseline Models

```python
model = LogisticRegression()
model = RandomForestClassifier(max_depth=1)
#model = SVC(C=0.6)
```

```python

X = df.drop(columns=["subject_id", "site_name", "diagnosis", "Gender", "Age"])
y = df["diagnosis"]

print(X.to_numpy().shape)
validate(model, X, y=="ADHD", metrics=["accuracy", "precision", "recall", "f1"])
```
